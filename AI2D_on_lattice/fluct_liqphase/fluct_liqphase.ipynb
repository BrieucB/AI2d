```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
list_beta=np.array([1.8, 2.0, 2.2, 2.4, 2.6, 2.8, 3.0])
```

```python
data_beta={}
for beta in list_beta:
    data_beta[beta]=np.loadtxt("profiles_beta/beta%.1f/f_profiles.dat"%beta)
```

```python
data_beta[3].shape
```

```python
for beta in list_beta:
    plt.matshow(data_beta[beta][-500:-1,:])
```

```python
fft=np.fft.fft2(data_beta[1.8][:500,:]-np.mean(data_beta[1.8]))
for k in range(1,21):
    fft+=np.fft.fft2(data_beta[1.8][500*k:500*(k+1),:]-np.mean(data_beta[1.8]))
fft/=22
```

```python
plt.plot(2*np.pi*np.fft.fftfreq(200, d=1)[:100], np.abs(fft)[0,:100], '-')
```

```python
plt.matshow(np.roll(np.abs(fft), axis=[0,1], shift=[250,250]))
```
