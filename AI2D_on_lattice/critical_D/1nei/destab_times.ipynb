# Destab times 

```python
list_D=[0.06, 0.08, 0.10, 0.12, 0.14, 0.16, 0.18, 0.20]
```

```python
t_destab={}
mean_t=[]
std_t=[]

for D in list_D:
    t_destab[D]=np.loadtxt('t_destab_D/D%s/f_td.dat'%str(D))
    mean_t.append(np.mean(t_destab[D]))
    std_t.append(np.std(t_destab[D]))
```

```python
plt.semilogy(list_D, mean_t, 'o')
```
