```python
td={}
for L in list_L:
    td[L]=np.array([])
#     for i in range(10):
#         try:
#             data=np.loadtxt("destab_D0.1_L/L%d/i%d/f_mag.dat"%(L,i), usecols=(0))[-1]
#             td[L].append(data)
#         except:
#             pass
    for i in range(10):
        try:
            data=np.loadtxt("destab_D0.1_L2/L%d/i%d/f_mag.dat"%(L,i), usecols=(0))
            td[L]=np.concatenate((td[L], data))
        except:
            pass

    for i in range(51):
        try:
            data=np.loadtxt("destab_D0.1_L_chihiro/L%d/i%d/f_mag.dat"%(L,i), usecols=(0))
            td[L]=np.concatenate((td[L], data))
        except:
            pass
```

```python
[len(td[tau]) for tau in td]
```

```python
a, b = np.polyfit(np.log(list(td)),
             np.log([np.mean(td[tau]) for tau in td]), deg=1)
```

```python
a, b,
```

```python
np.exp(20)
```

```python


plt.errorbar(list(td),
             [np.mean(td[tau]) for tau in td],
             yerr=3*np.array([np.std(td[tau]) for tau in td])/np.sqrt([len(td[tau]) for tau in td]),
             capsize=3,
             fmt='o')


plt.loglog(list_L, np.exp(b)*list_L**(a), 'k--')
ax=plt.gca()
ax.loglog()

plt.xticks([100, 200, 400, 800], ['%s'%a for a in [100, 200, 400, 800]])
plt.xlabel(r'$L$')
plt.ylabel(r'$\tau_d$')
```

```python
hist={}
bins = {}
for tau in td:
    hist[tau], bins[tau] = np.histogram(td[tau], bins=np.arange(0,1e5,50))
```
