```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
list_ly=np.arange(1, 11)
list_D=[0.2, 0.5, 0.7, 1, 1.2, 2]
```

```python
rev_t={}
mean_t={}

for D in list_D:
    rev_t[D]={}
    mean_t[D]=[]

    for ly in list_ly:
        try:
            rev_t[D][ly]=pd.read_csv('data_rev_ly_D%s/ly%d/f_td.dat'%(D,ly)).values[5:]
            print(D,ly, rev_t[D][ly].shape)
            mean_t[D].append(np.mean(rev_t[D][ly]))
            #std_t.append(np.std(rev_t[ly][::100]))
        except:
            pass
```

```python
for D in list_D:
    plt.semilogy(np.array(list(rev_t[D])), np.array(mean_t[D])/np.sqrt(D), 'o-', label=r'$D=%.2f$'%D)
plt.legend()
plt.xlabel(r'$l_y$')
plt.ylabel(r'$\langle \tau \rangle$')
```

```python
ly=4
[np.mean(rev_t[D][ly]) if ly in rev_t[D] else np.nan for D in list_D]
```

```python
for ly in list_ly[:6]:
    try:
        plt.semilogy(list_D, [np.mean(rev_t[D][ly]) if ly in rev_t[D] else np.nan for D in list_D], 'o-', label=ly)
    except:
        pass
plt.legend()
```

```python
ly=1
plt.plot(list_D, [np.mean(rev_t[D][ly]) if ly in rev_t[D] else np.nan for D in list_D], 'o-', label=ly)
```

```python
hist={}
bins={}

for D in list_D:
    hist[D]={}
    bins[D]={}
    for ly in list_ly:
        try:
            hist[D][ly], bins[D][ly]=np.histogram(rev_t[D][ly], bins=np.linspace(0,20000), density=True)
        except:
            pass
```

```python
plt.figure(figsize=(10,8))
D=1
for ly in hist[D]:
    plt.semilogy(bins[D][ly][:-1], hist[D][ly], 'o-', label='ly = %d'%ly)
plt.legend()
```
