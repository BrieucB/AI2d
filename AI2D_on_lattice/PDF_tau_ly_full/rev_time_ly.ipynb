```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
list_ly=np.arange(1, 11)
list_D=[0.2, 0.5, 0.7, 1, 1.2, 2]
```

```python
rev_t={}
mean_t={}

for D in list_D:
    rev_t[D]={}
    mean_t[D]=[]

    for ly in list_ly:
        print(D,ly)
        try:
            rev_t[D][ly]=pd.read_csv('data_rev_ly_D%s/ly%d/f_td.dat.gz'%(D,ly)).values[5:]
            print(D,ly, rev_t[D][ly].shape)
            mean_t[D].append(np.mean(rev_t[D][ly]))
            #std_t.append(np.std(rev_t[ly][::100]))
            
        except:
            try : 
                rev_t[D][ly]=pd.read_csv('data_rev_ly_D%s/ly%d/f_td.dat'%(D,ly)).values[5:]
                print(D,ly, rev_t[D][ly].shape)
                mean_t[D].append(np.mean(rev_t[D][ly]))
            except:pass
```

```python
hist={}
bins={}

for D in list_D:
    hist[D]={}
    bins[D]={}
    for ly in list_ly:
        try:
            hist[D][ly], bins[D][ly]=np.histogram(rev_t[D][ly], bins=np.linspace(0,20000), density=True)
        except:
            pass
```

```python
plt.figure(figsize=(10,8))
D=1
for ly in hist[D]:
    plt.semilogy(bins[D][ly][:-1]/(ly*ly), hist[D][ly], 'o-', label='ly = %d'%ly)
plt.legend()
```
