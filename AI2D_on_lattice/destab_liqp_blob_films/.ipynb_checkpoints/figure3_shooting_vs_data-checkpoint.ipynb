```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit, fixed_point
```

```python
from scipy.ndimage import uniform_filter
```

# $Âµ$ data

```python
list_beta=np.array([1.2, 1.4, 1.6, 1.8, 2.0, 2.2])
```

```python
lx=1000
```

```python
m_micro={}
for beta in list_beta:
    m_micro[beta]=pd.read_csv('data_speed_beta/beta%.1f/f_profiles_m.dat'%beta, sep=' ', names=np.arange(lx), index_col=False)
```

```python
k=7
beta=2.2
plt.matshow(-m_micro[beta].values[200*k:200*(k+1),:][:,::-1])
```

```python
beta=2.2
k=-2
prof=-m_micro[beta].values[200*k:200*(k+1),:][:,::-1]
avg_prof=uniform_filter(prof, size=6, mode='constant')

plt.plot(avg_prof[100,:])
```

```python
prof_mid=avg_prof[100,:]
np.where((prof_mid<-1)&(np.roll(prof_mid,-1)>=-1))
```

```python
posback={}
posfront={}
poscomet={}

for beta in list_beta:
    posback[beta]=[]
    posfront[beta]=[]
    poscomet[beta]=[]

    for k in range(1,7):
        prof=np.roll(-m_micro[beta].values[200*k:200*(k+1),:][:,::-1],0)
        avg_prof=uniform_filter(prof, size=6, mode='constant')
        prof_mid=avg_prof[100,:]
        poscomet[beta].append(np.where((prof_mid<-1)&(np.roll(prof_mid,-1)>=-1))[0][0])
        posback[beta].append(np.where((prof_mid<5)&(np.roll(prof_mid,-1)>=5))[0][0])
        posfront[beta].append(np.where((prof_mid>1)&(np.roll(prof_mid,-1)<=1))[0][0])
```

```python
speedback={}
speedfront={}
speedcomet={}
for beta in list_beta:
    speedfront[beta]=np.polyfit(50*np.arange(1,7), posfront[beta], deg=1)[0]
    speedback[beta]=np.polyfit(50*np.arange(1,7), posback[beta], deg=1)[0]
    speedcomet[beta]=np.polyfit(50*np.arange(1,7), poscomet[beta], deg=1)[0]
```

```python
plt.plot(list_beta, [speedfront[beta] for beta in list_beta], 'o')
plt.plot(list_beta, [speedback[beta] for beta in list_beta], 'x')
plt.plot(list_beta, [speedcomet[beta] for beta in list_beta], 'v')
```
