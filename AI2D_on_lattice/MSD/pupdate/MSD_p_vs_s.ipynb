```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
MSD_p=pd.read_csv('f_MSD.dat.gz', sep=' ', names=['t', 'MSD'])
MSD_s=pd.read_csv('../supdate/f_MSD.dat', sep=' ', names=['t', 'MSD'])
```

```python
plt.plot(MSD_p.t, MSD_p.MSD, '-', label='Parallel')
plt.plot(MSD_s.t, MSD_s.MSD, '-', label='Sequential')
plt.legend()
abs_t=np.linspace(0, np.max(MSD_s.t), 100)
D=1
d=2
plt.plot(abs_t, 2*d*D*abs_t, '--')
```

# $\beta = 0.9$,  $D = 1$

```python
list_v=[0.1, 0.5, 1.0, 2.0]
```

```python
dict_p={}
dict_s={}
for v in list_v:
    dict_p[v]=pd.read_csv('MSD_v0/v%s/f_MSD.dat'%v, sep=' ', names=['t', 'MSD'])
    dict_s[v]=pd.read_csv('../supdate/MSD_v0/v%s/f_MSD.dat'%v, 
                      sep=' ', names=['t', 'MSD'])
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label='Sequential',markevery=1000)
    plt.plot(dict_p[v].t, dict_p[v].MSD, 'v-', label='Parallel',markevery=1500)
plt.legend()
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label=r'$v=%s$'%v, markevery=1000)

abs_t=np.linspace(0, np.max(dict_s[v].t), 100)
D=1
v=2.16
plt.plot(abs_t, (4.*D+v)*abs_t, '--')
plt.legend()
```

# $\beta=2$, $D=0$

```python
list_v=[0.1, 0.5, 1.0, 2.0]
```

```python
dict_p={}
dict_s={}
for v in list_v:
    dict_p[v]=pd.read_csv('MSD_v2/v%s/f_MSD.dat'%v, sep=' ', names=['t', 'MSD'])
    dict_s[v]=pd.read_csv('../supdate/MSD_v2/v%s/f_MSD.dat'%v, 
                      sep=' ', names=['t', 'MSD'])
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label='Sequential v=%s'%v,markevery=2000)
    plt.plot(dict_p[v].t, dict_p[v].MSD, '-', label='Parallel v=%s'%v,markevery=2500)
plt.legend()

abs_t=np.linspace(0, np.max(dict_s[v].t), 1000)
#plt.plot(abs_t, 2.16*abs_t*abs_t, 'k--')
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label=r'$v=%s$'%v, markevery=1000)

abs_t=np.linspace(0, np.max(dict_s[v].t), 100)
D=1
v=2
plt.loglog(abs_t, 1500*abs_t, '--')
plt.legend()
```
