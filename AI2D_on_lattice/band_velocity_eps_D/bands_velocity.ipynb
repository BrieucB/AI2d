```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pandas as pd
```

```python
list_beta=[1.05, 1.1, 1.2, 1.4, 1.6, 1.9, 2.3, 2.5, 3.0]
```

```python
mag={}
for beta in list_beta:
    mag[beta]=np.loadtxt('instances_beta/beta%s/f_mid_mag.dat'%str(beta))
```

```python
for beta in list_beta:
    dt=(4+np.exp(beta))**(-1)
    plt.plot(dt*np.arange(0, mag[beta].shape[0]), mag[beta]-np.mean(mag[beta]), label=beta)
plt.legend()
```

# Crossing zero

```python
v=[]
for beta in list_beta:
    m=mag[beta]
    th=np.mean(m)
    t0=np.where((m>=th)&(np.roll(m,-1)<=th))
    dt=(4+np.exp(beta))**(-1)
    v.append(400/(dt*np.mean(np.diff(t0))))
```

```python
v
```

```python
plt.plot(1/np.array(list_beta), np.array(v)/1.8, 'o')
```
