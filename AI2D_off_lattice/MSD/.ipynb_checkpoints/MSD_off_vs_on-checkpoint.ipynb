# Turn on $v$, $\beta=2$

```python
list_v=[0.1, 0.5, 1.0, 2.0]
```

```python
dict_p={}
dict_s={}
for v in list_v:
    dict_p[v]=pd.read_csv('MSD_v1/v%s/f_MSD.dat'%v, sep=' ', names=['t', 'MSD'])
    dict_s[v]=pd.read_csv('../supdate/MSD_v1/v%s/f_MSD.dat'%v, 
                      sep=' ', names=['t', 'MSD'])
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label='Sequential',markevery=2000)
    plt.plot(dict_p[v].t, dict_p[v].MSD, '-', label='Parallel',markevery=2500)
plt.legend()

abs_t=np.linspace(0, np.max(dict_s[v].t), 1000)
plt.plot(abs_t, 2.16*abs_t*abs_t, 'k--')
```

```python
for v in list_v:
    plt.plot(dict_s[v].t, dict_s[v].MSD, 'x-', label=r'$v=%s$'%v, markevery=1000)

abs_t=np.linspace(0, np.max(dict_s[v].t), 100)
D=1
v=2
plt.plot(abs_t, (4.*D+v)*abs_t, '--')
plt.legend()
```
