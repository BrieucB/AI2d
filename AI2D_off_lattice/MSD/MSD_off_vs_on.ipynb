# $D = 1$,  $\beta = 0.9$

```python
list_v=[0.1, 0.5, 1.0, 2.0]
```

```python
dict_off={}
dict_on={}
for v in list_v:
    dict_off[v]=pd.read_csv('MSD_v0/v%s/f_MSD.dat'%v, sep=' ', names=['t', 'MSD'])
    dict_on[v]=pd.read_csv('../../AI2D_on_lattice/MSD/pupdate/MSD_v0/v%s/f_MSD.dat'%v, 
                      sep=' ', names=['t', 'MSD'])
```

```python
for v in list_v:
    plt.plot(dict_off[v].t, dict_off[v].MSD, 'x-', label='Off',markevery=10000)
    plt.plot(dict_on[v].t, dict_on[v].MSD, 'v-', label='On',markevery=1500)
plt.legend()
```

```python
for v in list_v:
    plt.plot(dict_off[v].t, dict_off[v].MSD, 'x-', label=r'$v=%s$'%v, markevery=10000)

abs_t=np.linspace(0, np.max(dict_off[v].t), 100)
D=1
v=1
plt.loglog(abs_t, (4.*D+v)*abs_t, '--')
plt.legend()
```
