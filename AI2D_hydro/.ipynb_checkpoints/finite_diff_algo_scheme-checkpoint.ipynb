```python
def solve_HDE_asters(L, dx, tmax, dt, D, v, beta, rho0):
    
    # Space and time discretization
    Nx=np.int64(L/dx)
    Nx2=np.int64(Nx/2)
    vec_t=np.arange(0,tmax,dt)
    
    # Coefficients
    µ=D*dt/(dx*dx)
    nu=v*dt/(dx)
    
    # Increment matrix
    lineID=np.array([1+2.*µ, -µ]+[0 for k in range(Nx-3)]+[-µ])
    IDbloc=np.array([np.roll(lineID, k) for k in range(Nx)])
    lineMix=np.array([0, nu]+[0 for k in range(Nx-3)]+[-nu])
    Mixbloc=np.array([np.roll(lineMix, k) for k in range(Nx)])
    A=np.block([[IDbloc, Mixbloc], [Mixbloc,IDbloc]])
    
    # Initial condition
    lam=np.log((2.*D+v)/(2.*D-v))
    rho=np.zeros(Nx)
    m=np.zeros(Nx)


    x=np.arange(Nx2+1, Nx)
    rho[Nx2+1:]=rho0*np.exp(-lam*L*(x-(Nx2+1))/Nx)
    m[Nx2+1:]=-rho0*np.exp(-lam*L*(x-(Nx2+1))/Nx)

    x=np.arange(0, Nx2+1)
    rho[:Nx2+1]=rho0*np.exp((-lam*L*(Nx2-x))/Nx)
    m[:Nx2+1]=rho0*np.exp((-lam*L*(Nx2-x))/Nx)

    #m+=1e-5*np.random.randn(Nx)
    X=np.concatenate([rho,m])
    
    # Observables
    M=[]
    res_rho=[]
    res_m=[]
    res_rho.append(rho)
    res_m.append(m)
    
    # Dynamic loop
    for t in vec_t:
        try:
            betaphi=np.divide(m, rho, where=rho!=0, out=np.zeros_like(m))
            b=np.concatenate([np.zeros_like(rho), 2*dt*rho*np.sinh(betaphi)-2*dt*m*np.cosh(betaphi)])
            X0=X.copy()
            X=np.linalg.solve(A, X0+b)
            rho, m = X[:Nx], X[Nx:]
            res_rho.append(rho.copy())
            res_m.append(m.copy())

            M.append(np.sum(m))
            #print(t, np.sum(m))
        except:
            break
        
    return(np.array(res_rho), np.array(res_m), M)
    
```

```python
def dXdt(X, L, dt, dx,N):
    rho=X[:L]
    theta=X[L:]
    
    drho=rho-(dt/dx)*(np.roll(theta*rho,  shift=-1)-np.roll(theta*rho,  shift=1))/2 \
    + 0.5*(dt/(dx*dx))*(np.roll(rho,  shift=-1)+np.roll(rho,  shift=1)-2*rho)
    
    dtheta=theta-(dt/dx)*(np.roll(np.log(rho),  shift=-1)-np.roll(np.log(rho),  shift=1))/2 \
    + (dt/(dx*dx))/(2*rho)*(np.roll(theta*rho,  shift=-1)+np.roll(theta*rho,  shift=1)-2*theta*rho) \
    - theta*drho/(rho) \
    + np.sqrt(2/(N*rho)*(1-theta*theta) + 4*rho*(np.sinh(beta*N*theta)+theta*np.cosh(beta*N*theta))/N)*np.random.rand(L)*np.sqrt(dt)
    
    dtheta[np.abs(dtheta)>1]=np.sign(dtheta[np.abs(dtheta)>1])
    
    return(np.concatenate([drho, dtheta]))
```

```python
L=2000
dt=1e-1
dx=1
N=2000
beta=2

vec_t=np.arange(0,10000, dt)

rho= (1/L)*np.ones(L)#np.concatenate([[0.3], 0.2*np.ones((L-1))])
theta=0*np.ones(L)#np.concatenate([[0.1], 0*np.ones((L-1))])
X=np.concatenate([rho, theta])
```

```python
prof_rho=[]
prof_theta=[]
prof_rho.append(rho.copy())
prof_theta.append(theta.copy())

for t in vec_t:
    X0=X.copy()
    
    X=dXdt(X0, L, dt, dx,N)
    
    rho, theta = X[:L], X[L:]  
    prof_rho.append(rho)
    prof_theta.append(theta)
```

```python
L=np.array([[0,1,0], [1, -4,1], [0,1,0]])
```

```python
L=np.array([[1/4,1/2,1/4], [1/2, -3,1/2], [1/4,1/2,1/4]])
```

```python
def update_rho(rho, m, v, D, dt, dx):
    #rho[rho<1e-3]=0
    return(rho 
           - v*(dt/dx)*(np.roll(m, axis=1,shift=-1)-np.roll(m, axis=1,shift=1))/2
           + D*(dt/(dx*dx))*convolve2d(rho, L, mode='same', boundary='wrap'))
```

```python
def update_m(rho, m, v, D, beta, dt, dx):
    #rho[rho<1e-3]=0    
    return(m 
           - v*(dt/dx)*(np.roll(rho, axis=1,shift=-1)-np.roll(rho, axis=1,shift=1))/2
           + D*(dt/(dx*dx))*convolve2d(m, L, mode='same', boundary='wrap')
           + 2*dt*(beta-1-1/rho)*m 
           #+ 2*dt*(beta-1-np.divide(1, rho, where=rho!=0, out=np.zeros_like(rho)))*m 
           - dt*beta*beta*(1-beta/3)*m*m*m/(rho*rho)
          )
```

```python
def update_X(X, v, D, beta, dt, dx, ly):
    rho, m = X[:ly, :], X[ly:, :]   
    
    drho = rho - v*(dt/dx)*(np.roll(m, axis=1,shift=-1)-np.roll(m, axis=1,shift=1))/2 + D*(dt/(dx*dx))*convolve2d(rho, L, mode='same', boundary='wrap')
    
    dm = m - v*(dt/dx)*(np.roll(rho, axis=1,shift=-1)-np.roll(rho, axis=1,shift=1))/2 + D*(dt/(dx*dx))*convolve2d(m, L, mode='same', boundary='wrap') + 2*dt*(beta-1-1/rho)*m- dt*beta*beta*(1-beta/3)*m*m*m/(rho*rho) 
       
    return(np.concatenate([drho, dm]))
```

```python
def outputCelerity(rho0, m0, tmax, dt, lx, ly, ds, v, beta, D):
    vec_t=np.arange(0,tmax,dt)
    X=np.concatenate([rho0, m0])
    pos_band=[]
    
    for t in vec_t:
        X0=X.copy()
        X=update_X(X0, v, D, beta, dt, ds, ly)
    
        rho, m = X[:ly, :], X[ly:, :]
        band_m=np.mean(m, axis=0)
        th=np.max(band_m/2)
        pos_band.append(np.where((band_m<th)&(np.roll(band_m, -1)>th))[0][0])
    
    return(np.polyfit(vec_t, pos_band, 1)[0])
```

```python
def update_Xk(X, v, D, beta, dt, dx, ly, l):
    rho, m = X[:ly, :], X[ly:, :]   
    
    Laprho=convolve2d(rho, L, mode='same', boundary='wrap')
    Lapm=convolve2d(m, L, mode='same', boundary='wrap')
    Gradrho=(np.roll(rho, axis=1,shift=-1)-np.roll(rho, axis=1,shift=1))/2
    Gradm=(np.roll(m, axis=1,shift=-1)-np.roll(m, axis=1,shift=1))/2
    ScaledLaprho=Laprho/rho
    
    drho = rho - \
    v*(dt/dx)*Gradm \
    + D*(dt/(dx*dx))*Laprho
    
    dm = m - \
    v*(dt/dxa*Gradrho \
    + (D+2*beta*l*l+ beta*beta*(beta-2)*l*l*m*m/(rho*rho))*(dt/(dx*dx))*Lapm \
    + dt*2*m*(beta*(1- l*l*ScaledLaprho/(dx*dx)) - 1)\
    - dt*(m*m*m/(rho*rho))*beta*beta*((1-beta/3) + l*l*(beta-2)*ScaledLaprho/(dx*dx)) \
    
    return(np.concatenate([drho, dm]))
```

```python
def update_Xk_RFMF(X, v, D, beta, dt, dx, ly, l):
    rho, m = X[:ly, :], X[ly:, :]   
    
    Laprho=convolve2d(rho, L, mode='same', boundary='wrap')
    Lapm=convolve2d(m, L, mode='same', boundary='wrap')
    Gradrho=(np.roll(rho, axis=1,shift=-1)-np.roll(rho, axis=1,shift=1))/2
    Gradm=(np.roll(m, axis=1,shift=-1)-np.roll(m, axis=1,shift=1))/2
    ScaledLaprho=Laprho/rho
    
    drho = rho - \
    v*(dt/dx)*Gradm \
    + D*(dt/(dx*dx))*Laprho
    
    dm = m - \
    v*(dt/dx)*Gradrho \
    + (D+2*beta*l*l+ beta*beta*(beta-2)*l*l*m*m/(rho*rho))*(dt/(dx*dx))*Lapm \
    + dt*2*m*(beta*(1- l*l*ScaledLaprho/(dx*dx)) - 1 - 1/rho)\
    - dt*(m*m*m/(rho*rho))*beta*beta*((1-beta/3) + l*l*(beta-2)*ScaledLaprho/(dx*dx)) \
    
    return(np.concatenate([drho, dm]))
```

# Crank-Nicolson for 2D heat equation

# Implicit x + explicit y - RFMF AI2D

```python
def RHS_rho(rho, m, µ, nu, dt, beta):
    return(rho*(1-2*µ)
           + 0.5*µ*(np.roll(rho, axis=0, shift=-1)
                   +np.roll(rho, axis=0, shift=1)
                   +np.roll(rho, axis=1, shift=-1)
                   +np.roll(rho, axis=1, shift=1))
           - nu*(np.roll(m, axis=0, shift=-1)
                   -np.roll(m, axis=0, shift=1))
          )
```

```python
def RHS_m(rho, m, µ, nu, dt, beta):
    return(m*(1-2*µ)
           + 0.5*µ*(np.roll(m, axis=0, shift=-1)
                   +np.roll(m, axis=0, shift=1)
                   +np.roll(m, axis=1, shift=-1)
                   +np.roll(m, axis=1, shift=1))
           - nu*(np.roll(rho, axis=0, shift=-1)
                   -np.roll(rho, axis=0, shift=1))
           + 2*dt*m*(beta-1-1/rho)
           -dt*beta*beta*(1-beta/3)*m*m*m/(rho*rho)
           
          )
```

```python
def updateOP(lx, ly, µ, nu):
    A1=np.block([[(1+µ)*np.eye(lx), np.zeros((lx,lx))],
         [np.zeros((lx,lx)), (1+µ)*np.eye(lx)]])
    
    mat2=np.zeros((lx, lx))
    for i in range(1, lx-1):
        mat2[i, i+1]=1
        mat2[i, i-1]=1

    mat2[0,1]=1
    mat2[-1, -2]=1
    mat2[0,-1]=1
    mat2[-1,0]=1

    A2=np.block([[-0.5*µ*mat2, np.zeros((lx,lx))],
             [np.zeros((lx,lx)), -0.5*µ*mat2]])
    
    mat3=np.zeros((lx, lx))
    for i in range(1, lx-1):
        mat3[i, i+1]=1
        mat3[i, i-1]=-1

    mat3[0,1]=1
    mat3[-1, -2]=-1
    mat3[0,-1]=-1
    mat3[-1,0]=1

    A3=np.block([[np.zeros((lx,lx)), nu*mat3],
             [nu*mat3, np.zeros((lx,lx))]])
    
    A=A1+A2+A3
    
    return(np.linalg.inv(A))
```
