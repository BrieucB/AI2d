# Runtimes

```python
nb_cores=np.array([6, 12, 18, 24, 30, 36, 42, 48, 54, 60, 64])
rt_omp=np.array([96.504, 49.083, 36.708, 27.921, 23.187, 20.803, 18.720, 18.021, 14.863, 17.375, 14.869])
rt_mpi=np.array([np.nan, 49.001, 33.940, 26.261, 21.543, 19.422, 18.120, 14.435, 13.392,12.926, 11.738])
```

```python
plt.plot(nb_cores, rt_omp/rt_mpi, '-o')
plt.xlabel("ncpu")
plt.ylabel(r"$\dfrac{time(openmp)}{time(mpi)}$", rotation=0, labelpad=40)
```

```python
plt.plot(nb_cores, rt_omp/nb_cores, '-o')
plt.xlabel("ncpu")
plt.ylabel(r"$\dfrac{time(openmp)}{time(mpi)}$", rotation=0, labelpad=40)
```

```python
plt.plot(nb_cores, rt_mpi*nb_cores, '-o', label=r"$time(mpi)/ncpu$")
plt.plot(nb_cores, rt_omp*nb_cores, '-x', label=r"$time(openmp)/ncpu$")

plt.legend()
#plt.ylim(0,5)
plt.xlabel("ncpu")
```
